# Engaged Vs Non Engaged Sessions

```python
session_df = df.groupby("channel group")[["Sessions", "Engaged Sessions"]].sum().reset_index()

# Non-Engaged sessions column 
session_df["Non-Engaged"] = session_df["Sessions"] - session_df["Engaged Sessions"]


session_df_melted = session_df.melt(
    id_vars="channel group",
    value_vars=["Engaged Sessions", "Non-Engaged"],
    var_name="Session Type",
    value_name="Count"
)

# Plot
plt.figure(figsize=(8,5))
sns.barplot(
    data=session_df_melted,
    x="channel group",
    y="Count",
    hue="Session Type"
)
plt.title("Engaged vs Non-Engaged Sessions")
plt.xticks(rotation=45)
plt.show()
```

# Traffic by Hour and Channel

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure DateHour is datetime
df["DateHour"] = pd.to_datetime(df["DateHour"])

# Hour 
df["Hour"] = df["DateHour"].dt.hour

# Pivot table: channel group x hour
traffic_pivot = df.pivot_table(
    index="channel group",
    columns="Hour",
    values="Sessions", 
    aggfunc="sum",
    fill_value=0
)

# Heatmap plot
plt.figure(figsize=(12,6))
sns.heatmap(
    traffic_pivot,
    cmap="viridis",   
    annot=True,
    fmt="d"
)
plt.title("Traffic by Hour and Channel Group")
plt.ylabel("Channel Group")
plt.xlabel("Hour of Day")
plt.show()
```

#Engagement rate Vs Sessions over Time

```python
import matplotlib.pyplot as plt

# make sure DateHour is datetime
df["DateHour"] = pd.to_datetime(df["DateHour"])

# Aggregate data by time 
time_df = df.groupby("DateHour")[["Engagement rate", "Sessions"]].mean().reset_index()

# Plot
fig, ax1 = plt.subplots(figsize=(12,6))

# Sessions (bar)
ax1.bar(time_df["DateHour"], time_df["Sessions"], color="skyblue", alpha=0.6, label="Sessions")
ax1.set_ylabel("Sessions", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")

# Engagement rate (line)
ax2 = ax1.twinx()
ax2.plot(time_df["DateHour"], time_df["Engagement rate"], color="red", marker="o", label="Engagement Rate")
ax2.set_ylabel("Engagement Rate (%)", color="red")
ax2.tick_params(axis="y", labelcolor="red")

# Title + formatting
plt.title("Engagement Rate vs Sessions Over Time")
fig.autofmt_xdate(rotation=45)
plt.show()

```
